```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "Coronavirus-Dataset/patient.csv"
coronavirus_data_df = pd.read_csv(csv_file)
coronavirus_data_df
```

```python
new_coronavirus_data_df = coronavirus_data_df.drop("disease", axis = 1)
new_coronavirus_data_df
```

```python
#df['elderly'] = np.where(df['age']>=50, 'yes', 'no')

#new_coronavirus_data_df["Visited Wuhan"] = np.where(new_coronavirus_data_df["infection_reason"]== "visit to Wuhan", 'true', 'false')

new_coronavirus_data_df["Visited Wuhan"] = np.where(new_coronavirus_data_df["infection_reason"].str.contains("Wuhan", na=False), 'true', 'false')  

new_coronavirus_data_df
```

```python
# Create database connection

connection_string = "postgres:postgres@localhost:5432/Coronavirus"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check table names

engine.table_names()
```

```python
new_coronavirus_data_df.to_sql(name='coronavirus_data', con=engine, if_exists='replace', index=False)
```
